```python
import pandas as pd 
```

```python
df = pd.read_csv(r'C:\Users\ASF\OneDrive\Desktop\project_dashpord\car_prices_jordan.csv')
df.head()
```

```python
df.info()
```

```python
df['brand']=df['Model'].apply(lambda x: x.split(' ')[0])
df['Year']=df['Model'].apply(lambda x: x.split(' ')[-1])
df['Name_Model']=df['Model'].apply(lambda x: ' '.join(x.split(' ')[1:-1]))

df[['brand','Year','Name_Model']].head()
```

```python
df['Power_CC']=df['Power'].str.extract(('(\d+)')).astype(float)
```

```python
df[['Power_CC','Power']].head()
```

```python
df['Price'].head()
```

```python
df['Price_JD']=df['Price'].astype(str).str.replace(',','')
```

```python
df['Price_JD']=pd.to_numeric(df['Price_JD'])
```

```python
print(df[['Price', 'Price_JD']].head())
```

```python
df['Price_JD'].info()
```

```python
df['Year'].dtype
```

```python
df['Year'] = df['Year'].astype(str).str.replace(r'\D', '', regex=True)
df['Year']=pd.to_numeric(df['Year']).astype('Int64')
```

```python
print(df['Year'].dtype)
```

```python
df.head()
```

```python
def clean_property(prop):
    prop_lower = str(prop).lower()
    
    if 'automatic' in prop_lower:
        return 'Automatic'  
    elif 'a/t' in prop_lower:
        return 'Automatic'
    elif 'manual' in prop_lower:
        return 'Manual'
    else:
        return prop 

df['Property'] = df['Property'].apply(clean_property)
```

```python
Year_mean = df['Year'].mean()

Y_rounded = round(Year_mean)
df['Year'] = df['Year'].fillna(Y_rounded)
```
